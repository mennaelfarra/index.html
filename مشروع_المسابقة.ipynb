````python
# Gaza Air Pollution Analysis using Google Earth Engine

# This script analyzes air pollutants over the Gaza Strip using satellite data
# from Sentinel-5P. The goal is to calculate the average concentration of
# major atmospheric pollutants between 2022 and 2025.

# The analysis focuses on the following gases:
# • Nitrogen Dioxide (NO2)
# • Sulfur Dioxide (SO2)
# • Carbon Monoxide (CO)

# -------------------------------------------------------

# 1️⃣ Study Area (AOI)
# The Area of Interest is the Gaza Strip boundary stored
# as a FeatureCollection in Google Earth Engine assets.

# Asset:
# projects/superb-leaf-461011-c9/assets/Gaza_Boundary_WGS84

# The map will center on the Gaza Strip and display its boundary.

# -------------------------------------------------------

# 2️⃣ Time Period
# The analysis covers a multi-year period:

# Start Date : 2022-01-01
# End Date   : 2025-12-31

# This allows comparison of pollution levels across several years.

# -------------------------------------------------------

# 3️⃣ Satellite Dataset

# Data Source:
# Sentinel-5P TROPOMI atmospheric monitoring satellite

# Dataset used in Earth Engine:
# COPERNICUS/S5P/OFFL

# The dataset provides global measurements of atmospheric gases
# including NO2, SO2, and CO.

# -------------------------------------------------------

# 4️⃣ Nitrogen Dioxide (NO2)

# Steps:
# • Load NO2 image collection
# • Filter images by the selected time range
# • Select the tropospheric NO2 band
# • Calculate the mean concentration
# • Clip the result to Gaza boundary
# • Display it on the map with a color scale

# Color palette represents concentration levels
# from low (blue) to high (red).

# -------------------------------------------------------

# 5️⃣ Sulfur Dioxide (SO2)

# Steps:
# • Load SO2 satellite dataset
# • Filter by date range
# • Select SO2 column density band
# • Calculate the average value
# • Clip to Gaza boundary
# • Visualize it using a color gradient

# -------------------------------------------------------

# 6️⃣ Carbon Monoxide (CO)

# Steps:
# • Load CO dataset from Sentinel-5P
# • Filter by selected period
# • Calculate mean concentration
# • Clip the result to Gaza area
# • Display on the map with visualization parameters

# -------------------------------------------------------

# 7️⃣ Visualization

# Each pollutant layer is added to the map using:
# Map.addLayer()

# Visualization parameters include:
# • Minimum value
# • Maximum value
# • Color palette representing pollution intensity

# This allows visual comparison of pollution levels across Gaza.

# -------------------------------------------------------

# Output:
# Interactive map showing average concentrations of
# NO2, SO2, and CO over the Gaza Strip between 2022-2025.
````

````python
# Gaza Air Pollution Monitoring (Automated Pipeline)

# This repo automates **daily monitoring of air pollutants over Gaza** using Google Earth Engine (GEE) + GitHub Actions.

# Exports **time-series CSVs** and **GeoTIFF maps** to Google Drive, and saves an interactive HTML map locally.

# * **NO2 (nitrogen dioxide)**
# * **SO2 (sulfur dioxide)**
# * **CO (carbon monoxide)**
# * **AOD (aerosol optical depth)**
# * **UVAI (ultraviolet aerosol index)**

# ## Features

# * Automated via **GitHub Actions** (runs daily)
# * Auth via **Service Account JSON** (CI-friendly)
# * Outputs **GeoTIFFs** to **Google Drive**
# * Telegram notifications for success/failure
# * Daily **annotated GeoTIFF maps** with title/date + colorbar (for visualization)
# * Exports in **GeoTIFF** format (KML is not supported for Drive export)
# * Composite overlays with **FIRMS fires** + **ERA5 wind vectors**
# * Saves **interactive HTML map** locally

# ---

# ## Export CSVs (exporter.py)

# [unchanged from previous section, still exports pollutant CSVs]

# ---

# ## Export GeoTIFF Maps to Google Drive (render_latest_maps.py)

import os, json, datetime as dt
import ee
import matplotlib.pyplot as plt
import numpy as np

# Auth
sa_key_json = os.environ.get('EE_SA_KEY')
if sa_key_json:
    svc = json.loads(sa_key_json)
    credentials = ee.ServiceAccountCredentials(svc['client_email'], key_data=sa_key_json)
    ee.Initialize(credentials, project='superb-leaf-461011-c9') # Add project ID here
elif os.path.exists('ee_sa_key.json'): # Check for local key file
    with open('ee_sa_key.json') as f:
        svc = json.load(f)
    credentials = ee.ServiceAccountCredentials(svc['client_email'], key_data=json.dumps(svc))
    ee.Initialize(credentials, project='superb-leaf-461011-c9') # Add project ID here
else:
    # Authenticate Earth Engine in Colab
    ee.Authenticate()
    ee.Initialize(project='superb-leaf-461011-c9') # Add project ID here

# Params
AOI = ee.Geometry.Rectangle([34.20, 31.20, 34.63, 31.61], None, False)
END = dt.date.today().isoformat()
# BUCKET is no longer used, exports go to Google Drive
DRIVE_FOLDER = "gaza_air_quality_exports" # Define the Google Drive folder name for GeoTIFF exports

# Collections
s5pNO2 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_NO2').filterDate('2024-01-01', END)
s5pSO2 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_SO2').filterDate('2024-01-01', END)
s5pCO  = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_CO').filterDate('2024-01-01', END)
s5pAER = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_AER_AI').filterDate('2024-01-01', END)
maiac  = ee.ImageCollection('MODIS/061/MCD19A2').filterDate('2024-01-01', END)
firms  = ee.ImageCollection('FIRMS').filterDate(END, END)
era5   = ee.ImageCollection('ECMWF/ERA5/DAILY').filterDate(END, END)

# Latest images with QA
latest = {
  'NO2': s5pNO2.sort('system:time_start', False).first()
            .updateMask(ee.Image(s5pNO2.first()).select('qa_value').gte(0.75))
            .select('NO2_column_number_density').rename('no2'),
  'SO2': s5pSO2.sort('system:time_start', False).first()
            .updateMask(ee.Image(s5pSO2.first()).select('qa_value').gte(0.5))
            .select('SO2_column_number_density').rename('so2'),
  'CO' : s5pCO.sort('system:time_start', False).first()
            .updateMask(ee.Image(s5pCO.first()).select('qa_value').gte(0.5))
            .select('CO_column_number_density').rename('co'),
  'UVAI': s5pAER.sort('system:time_start', False).first()
            .select('absorbing_aerosol_index').rename('uv_ai'),
  'AOD': maiac.sort('system:time_start', False).first()
            .select('Optical_Depth_055').multiply(0.001).rename('aod') if maiac.sort('system:time_start', False).first() else None,
}

# Visualization ranges
vis = {
  'NO2': {'min': 0, 'max': 2e-4, 'palette': ['black','purple','blue','cyan','green','yellow','orange','red']},
  'SO2': {'min': 0, 'max': 5e-5, 'palette': ['black','blue','cyan','green','yellow','orange','red']},
  'CO' : {'min': 0, 'max': 0.05,  'palette': ['black','navy','blue','cyan','green','yellow','orange','red']},
  'UVAI': {'min': -2, 'max': 3,   'palette': ['#2c7bb6','#abd9e9','#ffffbf','#fdae61','#d7191c']},
  'AOD': {'min': 0, 'max': 1.0,   'palette': ['white','yellow','orange','red','maroon']},
}

# Export loop
for name, img in latest.items():
    if img is None:
        print(f"Skipping export for {name} as no data is available.")
        continue

    img = img.clip(AOI)

    # GeoTIFF export with annotation + colorbar to Google Drive
    desc = f"{name}_latest_geotiff"
    # Use a simple file name prefix for Drive
    fileNamePrefix_annotated = f"{name}_latest_annotated_{END}"
    task = ee.batch.Export.image.toDrive(
        image=img.visualize(**vis[name])
              .paint(firms.mosaic(), 1, 2)  # overlay FIRMS
              .blend(era5.first().select('u_component_of_wind_10m')),
        description=desc,
        folder=DRIVE_FOLDER, # Export to Google Drive folder
        fileNamePrefix=fileNamePrefix_annotated,
        fileFormat='GeoTIFF',
        scale=2000,
        region=AOI,
        maxPixels=1e13
    )
    task.start()

    # GeoTIFF export to Google Drive
    desc_tif = f"{name}_geotiff"
    # Use a simple file name prefix for Drive
    fileNamePrefix_tif = f"{name}_{END}"
    task_tif = ee.batch.Export.image.toDrive(
        image=img,
        description=desc_tif,
        folder=DRIVE_FOLDER, # Export to Google Drive folder
        fileNamePrefix=fileNamePrefix_tif,
        fileFormat='GeoTIFF',
        scale=2000,
        region=AOI,
        maxPixels=1e13
    )
    task_tif.start()

    # KML export - Removed as it's not supported for image export to Drive
    # task_kml = ee.batch.Export.image.toCloudStorage(
    #     image=img.visualize(**vis[name]),
    #     description=f"{name}_kml",
    #     bucket=BUCKET,
    #     fileNamePrefix=f"{PREFIX}/{name}.kml",
    #     fileFormat='KML',
    #     scale=2000,
    #     region=AOI,
    #     maxPixels=1e13
    # )
    # task_kml.start()

    print('Started Drive export for', name, task.id)

# ---

# ## GitHub Actions Update

# Add under the `exporter.py` step:

# ```yaml
#       - name: Export GeoTIFF maps to Drive
#         env:
#           EE_SA_KEY: ${{ secrets.EE_SA_KEY }}
#           # GCS_BUCKET is no longer needed for Drive export
#         run: python render_latest_maps.py
# ```

# ---

# ✅ Outputs exported to Google Drive:

# * Annotated **GeoTIFF** maps with title, date, colorbar
# * **GeoTIFF** rasters for scientific use
# * Overlays include **FIRMS fires** + **ERA5 wind**

# ---

# ## NEW: Interactive HTML Map (render_map_html.py)

# This script builds a `folium` interactive map with layer switcher, legends, title/date, pollutant tiles from Earth Engine, FIRMS fires as point markers, and ERA5 wind speed raster + sampled wind vectors (arrows). It saves `gaza_air_quality_<DATE>.html` locally.

import os, json, datetime as dt
import ee
import folium
from folium import Map, TileLayer, LayerControl, features
from branca.colormap import LinearColormap
# google.cloud.storage is not needed for Drive export or local HTML save

# Initialize Earth Engine with service account JSON stored in EE_SA_KEY
sa_key_json = os.environ.get('EE_SA_KEY')
if sa_key_json:
    svc = json.loads(sa_key_json)
    creds = ee.ServiceAccountCredentials(svc['client_email'], key_data=sa_key_json)
    ee.Initialize(creds, project='superb-leaf-461011-c9') # Add project ID here
elif os.path.exists('ee_sa_key.json'): # Check for local key file
    with open('ee_sa_key.json') as f:
        svc = json.load(f)
    creds = ee.ServiceAccountCredentials(svc['client_email'], key_data=json.dumps(svc))
    ee.Initialize(creds, project='superb-leaf-461011-c9') # Add project ID here
else:
    # Authenticate Earth Engine in Colab
    ee.Authenticate()
    ee.Initialize(project='superb-leaf-461011-c9') # Add project ID here

# Params
AOI = ee.Geometry.Rectangle([34.20, 31.20, 34.63, 31.61], None, False)
CENTER = [31.45, 34.40]  # [lat, lon]
TODAY = dt.date.today().isoformat()
# BUCKET and PREFIX are no longer used
# HTML file is saved locally

# Collections (use recent data)
s5pNO2 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_NO2').filterDate('2024-01-01', TODAY)
s5pSO2 = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_SO2').filterDate('2024-01-01', TODAY)
s5pCO  = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_CO').filterDate('2024-01-01', TODAY)
s5pAER = ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_AER_AI').filterDate('2024-01-01', TODAY)
maiac  = ee.ImageCollection('MODIS/061/MCD19A2').filterDate('2024-01-01', TODAY)

# FIRMS (Active fires) — attempt common collections
try:
    firms = ee.FeatureCollection('FIRMS')
except Exception:
    # fallback to MODIS/VIIRS near-real-time fire collections if FIRMS alias not present
    try:
        firms = ee.FeatureCollection('FIRMS/viirs')
    except Exception:
        firms = None

# ERA5 daily (u/v at 10m or nearest available daily product)
era5 = ee.ImageCollection('ECMWF/ERA5/DAILY').filterDate(TODAY, TODAY)

# Build latest pollutant images with simple QA
latest_images = {}
latest_images['NO2'] = s5pNO2.sort('system:time_start', False).first().select('NO2_column_number_density')
latest_images['SO2'] = s5pSO2.sort('system:time_start', False).first().select('SO2_column_number_density')
latest_images['CO']  = s5pCO.sort('system:time_start', False).first().select('CO_column_number_density')
latest_images['UVAI']= s5pAER.sort('system:time_start', False).first().select('absorbing_aerosol_index')
# Add check for AOD image before selecting band
aod_img = maiac.sort('system:time_start', False).first()
latest_images['AOD'] = aod_img.select('Optical_Depth_055').multiply(0.001) if aod_img else None


# Visualization params (same palettes as earlier, kept simple for web)
vis_params = {
    'NO2': {'min':0, 'max':2e-4, 'palette': ['black','purple','blue','cyan','green','yellow','orange','red']},
    'SO2': {'min':0, 'max':5e-5, 'palette': ['black','blue','cyan','green','yellow','orange','red']},
    'CO' : {'min':0, 'max':0.05,  'palette': ['black','navy','blue','cyan','green','yellow','orange','red']},
    'UVAI':{'min':-2, 'max':3,    'palette': ['#2c7bb6','#abd9e9','#ffffbf','#fdae61','#d7191c']},
    'AOD': {'min':0, 'max':1.0,   'palette': ['white','yellow','orange','red','maroon']}
}

# Create folium map
m = Map(location=CENTER, zoom_start=10, tiles='CartoDB positron')

# Add title (HTML)
title_html = f"<h3 align=\'center\' style=\'font-size:16px\'><b>Gaza Air Pollution — {TODAY}</b></h3>"
m.get_root().html.add_child(folium.Element(title_html))

# Function to add EE image as a folium TileLayer
def add_ee_layer(image, vis, name):
    map_id_dict = ee.Image(image).getMapId(vis)
    tile_url = map_id_dict['tile_fetcher'].url_format
    TileLayer(tiles=tile_url, attr='Google Earth Engine', name=name, overlay=True, control=True).add_to(m)

# Add pollutant layers
for name, img in latest_images.items():
    if img is not None: # Add check here before adding layer
        try:
            add_ee_layer(img.clip(AOI), vis_params[name], name)
        except Exception as e:
            print('Failed adding', name, e)
    else:
        print(f"No image data available for {name}, skipping layer.")


# Add FIRMS fires as a MarkerCluster (if available)
if firms is not None:
    try:
        today_fires = firms.filterDate(TODAY, ee.Date(TODAY).advance(1, 'day')).filterBounds(AOI)
        # sample up to 500 features to avoid massive JS
        fires_fc = today_fires.limit(500)
        fires = fires_fc.getInfo()
        fire_group = folium.FeatureGroup(name='FIRMS Fires', show=False)
        for f in fires['features']:
            coords = f['geometry']['coordinates']
            props = f.get('properties', {})
            popup = ', '.join([f"{k}: {v}" for k, v in props.items()])
            folium.CircleMarker(location=[coords[1], coords[0]], radius=4, color='red', fill=True, fillOpacity=0.8, popup=popup).add_to(fire_group)
        m.add_child(fire_group)
    except Exception as e:
        print('FIRMS layer failed:', e)

# Add ERA5 wind speed raster as a layer and sample vectors for arrows
try:
    era_img = era5.first()
    u = era_img.select('u_component_of_wind_10m')
    v = era_img.select('v_component_of_wind_10m')
    speed = u.hypot(v).rename('wind_speed')
    # add raster
    add_ee_layer(speed.clip(AOI), {'min':0, 'max':15, 'palette':['white','blue','green','yellow','red']}, 'ERA5 Wind Speed')

    # sample a coarse grid of points within AOI for arrows
    lon_min, lat_min, lon_max, lat_max = 34.20, 31.20, 34.63, 31.61
    lons = list(np.arange(lon_min, lon_max, 0.03))
    lats = list(np.arange(lat_min, lat_max, 0.03))
    points = []
    for lon in lons:
        for lat in lats:
            points.append(ee.Feature(ee.Geometry.Point([lon, lat])))
    pts_fc = ee.FeatureCollection(points)
    sampled = pts_fc.map(lambda f: f.set(u.reduceRegion(ee.Reducer.first(), f.geometry(), 10000)).set(v.reduceRegion(ee.Reducer.first(), f.geometry(), 10000)))
    sampled_info = sampled.limit(500).getInfo()
    arrow_group = folium.FeatureGroup(name='Wind Vectors', show=False)
    for f in sampled_info['features']:
        geom = f['geometry']
        props = f.get('properties', {})
        if props.get('u_component_of_wind_10m') is None or props.get('v_component_of_wind_10m') is None:
            continue
        uval = float(props['u_component_of_wind_10m'])
        vval = float(props['v_component_of_wind_10m'])
        # convert to a small line for arrow: compute end point in degrees (approx)
        lon, lat = geom['coordinates']
        scale = 0.02
        end_lat = lat + vval * scale
        end_lon = lon + uval * scale
        folium.PolyLine(locations=[[lat, lon], [end_lat, end_lon]], color='black', weight=1).add_to(arrow_group)
    m.add_child(arrow_group)
except Exception as e:
    print('ERA5 wind layer failed:', e)

# Legends: create simple colorbars with branca
for name, p in vis_params.items():
    cmap = LinearColormap(p['palette'], vmin=p['min'], vmax=p['max'])
    cmap.caption = name
    m.add_child(cmap)

# Layer control
LayerControl().add_to(m)

# Save HTML locally
out_html = f"gaza_air_quality_{TODAY}.html"
m.save(out_html)
print('Saved', out_html)

# HTML is saved locally, no upload to GCS or Drive from this script
# Use separate steps in your workflow to handle the local HTML file if needed

# ---

# ## Update GitHub Actions

# Add this step after the GeoTIFF exports:

# ```yaml
#       - name: Build interactive HTML map (saved locally)
#         env:
#           EE_SA_KEY: ${{ secrets.EE_SA_KEY }}
#           # GCS_BUCKET is not needed
#         run: |
#           pip install folium branca # Required for creating the map
#           # google-cloud-storage is not needed
#           python render_map_html.py
#       # Add subsequent steps here if you need to upload the generated HTML file
#       # For example, to GitHub Pages, a different cloud storage, or an artifact
# ```

# ---

# You're all set — this will create daily GeoTIFFs exported to your Google Drive folder and an interactive HTML map saved locally within the execution environment.

# If you'd like, I can also help with:

# * hosting the HTML on **GitHub Pages** automatically (requires additional workflow steps), or
# * packaging everything into a ready-to-run ZIP you can drop into a repo.
````

````python
# ============================================================
# Gaza Air Pollution Analysis (2022–2025)
# Google Earth Engine Python Script
# ============================================================

import ee
import geemap

# Authenticate and initialize Earth Engine
try:
    ee.Initialize(project='superb-leaf-461011-c9')
except:
    ee.Authenticate()
    ee.Initialize(project='superb-leaf-461011-c9')

# ------------------------------------------------------------
# AOI (Gaza Boundary)
# ------------------------------------------------------------

aoi = ee.FeatureCollection(
    'projects/superb-leaf-461011-c9/assets/Gaza_Boundary_WGS84'
)

# Create interactive map
Map = geemap.Map()

Map.centerObject(aoi, 9)
Map.addLayer(aoi, {'color': 'red'}, 'Gaza Boundary')

# ------------------------------------------------------------
# Date Range
# ------------------------------------------------------------

start = '2022-01-01'
end   = '2025-12-31'

# ------------------------------------------------------------
# NO2
# ------------------------------------------------------------

no2 = (
    ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_NO2')
    .filterDate(start, end)
    .select('tropospheric_NO2_column_number_density')
    .mean()
    .clip(aoi)
)

Map.addLayer(
    no2,
    {
        'min': 0.000001,
        'max': 0.00005,
        'palette': ['blue', 'green', 'yellow', 'orange', 'red']
    },
    'NO2 Gaza'
)

# ------------------------------------------------------------
# SO2
# ------------------------------------------------------------

so2 = (
    ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_SO2')
    .filterDate(start, end)
    .select('SO2_column_number_density')
    .mean()
    .clip(aoi)
)

Map.addLayer(
    so2,
    {
        'min': 0,
        'max': 0.0005,
        'palette': ['blue', 'yellow', 'red']
    },
    'SO2 Gaza'
)

# ------------------------------------------------------------
# CO
# ------------------------------------------------------------

co = (
    ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_CO')
    .filterDate(start, end)
    .select('CO_column_number_density')
    .mean()
    .clip(aoi)
)

Map.addLayer(
    co,
    {
        'min': 0,
        'max': 0.05,
        'palette': ['green', 'orange', 'red']
    },
    'CO Gaza'
)

# ------------------------------------------------------------
# UVAI
# ------------------------------------------------------------

uvai = (
    ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_AER_AI')
    .filterDate(start, end)
    .select('absorbing_aerosol_index')
    .mean()
    .clip(aoi)
)

Map.addLayer(
    uvai,
    {
        'min': -1,
        'max': 3,
        'palette': ['blue', 'white', 'red']
    },
    'UVAI Gaza'
)

# ------------------------------------------------------------
# AOD (Aerosol Height proxy dataset)
# ------------------------------------------------------------

aod = (
    ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_AER_LH')
    .filterDate(start, end)
    .select('aerosol_height')
    .mean()
    .clip(aoi)
)

Map.addLayer(
    aod,
    {
        'min': 0,
        'max': 3000,
        'palette': ['purple', 'yellow', 'red']
    },
    'AOD Gaza'
)

# ------------------------------------------------------------
# Display Map
# ------------------------------------------------------------

Map
````
